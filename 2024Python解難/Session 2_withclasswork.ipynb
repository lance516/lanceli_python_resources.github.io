```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
def getGGR(year):
    browser = webdriver.Chrome()
    browser.get(f'https://www.dicj.gov.mo/web/cn/information/DadosEstat_mensal/{year}/index.html')
    time.sleep(3)
    table = browser.find_element(By.ID, 'table1')
    table.get_attribute('outerHTML')
    data = pd.read_html(table.get_attribute('outerHTML'))[0]
    data = data.iloc[3:, 0:2]
    browser.quit()
    data.columns=['Month', 'GGR']
    data['Month'] = [i for i in range(1,13)]
    data['year'] = year
    data = data[data['GGR'] != '-']
    data['GGR'] = data['GGR'].astype('int')
    return data

result = pd.DataFrame()
for i in range(2018,2025):
    result = pd.concat([result, getGGR(i)])

display(result)
```

```python
import matplotlib.pyplot as plt
result['Date'] = result['year'].astype(str) + '-' + result['Month'].astype(str)
result['Date'] = pd.to_datetime(result['Date'], format='%Y-%m')
fig, ax = plt.subplots(figsize=(10,3))
result.plot(kind='line', x='Date', y='GGR', ax=ax)
```

```python
import untangle
import requests
def getGGR_XML(year):
    res = requests.get('https://www.dicj.gov.mo/web/cn/information/DadosEstat_mensal/2023/report_cn.xml')
    res.encoding='utf-8'
    year_data=pd.DataFrame()
    data_xml = untangle.parse(res.text)
    for i, d in enumerate(data_xml.STATISTICS.REPORT.DATA.RECORD):
        year_data = pd.concat([year_data, (pd.DataFrame([{'Month': i+1, 'GGR': d.DATA[1].cdata}]))], ignore_index=True)
    year_data['year']=year
    return year_data

result = pd.DataFrame()
for i in range(2018,2025):
    result = pd.concat([result, getGGR_XML(i)])
```

```python
import requests
import pandas as pd
import math
pd.set_option('display.max_rows', 500)

# Assume we are at below location (CPTTM)
latitude = 22.1924743
longitude = 113.5421817

headers = {
    'Authorization': 'APPCODE 09d43a591fba407fb862412970667de4'
}

car_park_data = requests.get('https://dsat.apigateway.data.gov.mo/car_park_maintance', headers=headers)
car_park_detail = requests.get('https://dsat.apigateway.data.gov.mo/car_park_detail', headers=headers)
car_park_data.encoding = 'utf-8'
car_park_detail.encoding = 'utf-8'

car_park_detail = pd.read_xml(car_park_detail.text)
car_park_data = pd.read_xml(car_park_data.text)

results = pd.merge(left=car_park_data, right=car_park_detail, left_on=['ID'],right_on=['CP_ID'])
results = results[results.Car_CNT>0]
results['Distance'] = ((results['X_coords'] - latitude)**2 + (results['Y_coords'] - longitude)**2)**0.5
results[['ID','name','X_coords','Y_coords','Distance','Car_CNT']].sort_values('Distance', ascending=True).head(5)
```
