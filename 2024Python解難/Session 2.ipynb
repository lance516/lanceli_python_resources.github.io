## Macau Airport Data



```python
import requests
import panadas as pd

headers = {
    'Authorization':'APPCODE 09d43a591fba407fb862412970667de4'
}
res = requests.get('https://aacm.apigateway.data.gov.mo/api/open/listFlightTimetable', headers=headers)
res.encoding='utf-8'
data = pd.read_json(res.text)
data
```

```python
data['periodStartDate'] = pd.to_datetime(data['periodStartDate'], format='%d-%m-%Y')
data['periodEndDate'] = pd.to_datetime(data['periodEndDate'], format='%d-%m-%Y')
data.info()
```

```python
data['weekDay'] = data['weekDay'].str.split(',')
data
```

```python
data2 = data.explode('weekDay')
data2['Date'] = data2.apply(
    lambda x: pd.date_range(x['periodStartDate'], x['periodEndDate'], freq='D'), axis=1
)
display(data2)
data3 = data2.explode('Date')
data3
```

```python
data3 = data3.reset_index()
```

```python
data3['new weekday'] = data3['Date'].dt.dayofweek + 1
data3['weekDay'] = data3['weekDay'].astype(int)
data3[data3['new weekday']==data3['weekDay']]
data4 = data3.drop(columns=['index','periodStartDate','periodEndDate','weekDay'])
data4
```
