```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
dfs = []
```

```python
# 費時佢死機，要分開攞
for page_index in (7,10):
    url = f"https://mo.centanet.com/Transaction?FileType=%E4%BD%8F%E5%AE%85&PropType=S&TitleName=2025%E5%B9%B4&BeginTxDate=2025-01-01&EndTxDate=2025-5-8&DateValue=2025&DataType=Y&PageIndex={page_index}"
    res = requests.get(url)
    tables = pd.read_html(res.text)
    dfs.append(tables[0])
    time.sleep(0.5)
```

```python
df = pd.concat(dfs)
df.reset_index(drop=True, inplace=True)

df
```
